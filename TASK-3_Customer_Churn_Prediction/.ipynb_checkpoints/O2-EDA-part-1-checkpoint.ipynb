## Surname

`The customer's surname or last name`

```python
def cat_eda(df, column):
    print(f"Null values: {df[column].isnull().sum()}")
    print("*"*20)
    value_counts = df[column].value_counts()
    print(value_counts)
    
    if df[column].nunique() > 30:
        fig, axes = plt.subplots(2, 1, figsize=(15, 10))
        df = df[column].value_counts().head(30).reset_index()
        sns.barplot(x=column, y='count', data=df, ax=axes[0])
        df = df.head(10)
        axes[1].pie(df['count'], labels=df[column], shadow=True, startangle=140)
    else:
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        sns.countplot(x=column, data=df, ax=axes[0])
        axes[1].pie(value_counts.values, labels=value_counts.index, shadow=True, startangle=140)
    plt.setp(axes[0].get_xticklabels(), rotation=90)
    plt.subplots_adjust(hspace=0.8)
```

```python
cat_eda(df, "Surname")
```

```python
def num_eda(column):
    print(f"Null values: {df[column].isnull().sum()}")
    print("*"*20)
    print(df[column].describe())
    df[column].hist(bins=20, figsize=(15, 6))

    fig, axes = plt.subplots(1, 2, figsize=(15, 4))
    sns.kdeplot(x=column, data=df, ax=axes[0])
    axes[0].set_title(f'Skewness: {df[column].skew()}')
    sns.boxplot(x=column, data=df, ax=axes[1])
```

```python
def get_outlier_range(column):
    perc_25 = np.percentile(df[column], 25)
    perc_75 = np.percentile(df[column], 75)
    iqr = perc_75 - perc_25

    upper_bound = perc_75 + (1.5 * iqr)
    lower_bound = perc_25 - (1.5 * iqr)

    return {
        "ub": upper_bound,
        "lb": lower_bound
    }
```

## CreditScore

`A numerical value representing the customer's credit score`

```python
num_eda("CreditScore")
```

```python
rng = get_outlier_range("CreditScore")
df[(df["CreditScore"] < rng['lb']) | (df["CreditScore"] > rng['ub'])]
```
