## Text Preprocessing and EDA

### Overview
In this file, I focused on text preprocessing and exploratory data analysis (EDA) for the SMS Spam Collection Dataset. The tasks included:

1. **Text Preprocessing**
   - Lowercasing the text.
   - Removing punctuation marks.

2. **Exploratory Data Analysis (EDA)**

## Cleaning text

```python
stemmer = PorterStemmer()
```

```python
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    words = nltk.word_tokenize(text)

    l = []
    for i in words:
        if i not in stopwords and len(i) > 2:
            l.append(i)
    return " ".join(l)

def remove_puctuations(text):
    l = []
    for i in text:
        if i not in string.punctuation:
            l.append(i)
    return "".join(l)


def preprocess_text(text):
    # Convert to lower case
    text = text.lower()
    
    # keep only english chars / remove numbers from ducuments
    text = re.sub(r'[^a-zA-Z+]', ' ', text)
    
    # Removing puctuations from all ducuments
    text = remove_puctuations(text)

    # Remove stop words
    text = remove_stopwords(text)

    # Removing repeated/leading/trailing spaces
    text = re.sub("\s[\s]+", " ",text).strip()

    # Tokenize
    text = nltk.word_tokenize(text)

    # Stemming
    words = []
    for word in text:
        words.append(stemmer.stem(word))

    return " ".join(words)
```

```python
tqdm.pandas(desc="text_preprocessing")
df["cleaned_text"] = df["text"].progress_apply(preprocess_text)
```

## Wordclouds

```python
from wordcloud import WordCloud
wc = WordCloud(width=500,height=500,min_font_size=10,background_color='white')
```

```python
spam_wc = wc.generate(df[df['target'] == "spam"]['cleaned_text'].str.cat(sep=" "))
plt.figure(figsize=(15,6))
plt.imshow(spam_wc)
```

```python
ham_wc = wc.generate(df[df['target'] == "ham"]['cleaned_text'].str.cat(sep=" "))
plt.figure(figsize=(15,6))
plt.imshow(ham_wc)
```
