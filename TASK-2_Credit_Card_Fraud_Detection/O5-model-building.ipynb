```python
import numpy as np
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("data/credit_card_fraud_detection_feature_engineering.csv")
df.sample(2)
```

```python
X = df.drop(columns=['is_fraud'])
y = df['is_fraud']
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

## Building Model

Testing **LogisticRegression**, **DecisionTree** and **RandomForest**

```python
models = {
    "log_reg": LogisticRegression(solver='lbfgs', max_iter=1000),
    "dt": DecisionTreeClassifier(),
    "rf": RandomForestClassifier()
}
```

```python
res = []

for model_name, model in models.items():
    # K-fold cross-validation
    kfold = KFold(n_splits=10, shuffle=True, random_state=42)
    scores = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')

    res.append([model_name, scores.mean(),scores.std()])

model_summary = pd.DataFrame(res, columns=['model', 'accuracy', 'accuracy_std'])
```

```python
model_summary
```

**RandomForest** gives the best mean accuracy

## Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Instantiate random forest model.
rf = RandomForestClassifier()
```

```python
# Determine set of hyperparameters for random forest
rf_cv_params = {'n_estimators' : [50,100 , 150 , 200],
              'max_depth':[4,5,6,7,8,9,10,11,12,15,20,30,40,50],
             'min_samples_leaf': [2, 5, 10, 20, 50],
             'min_samples_split': [2, 4, 6],
              'max_features' : ["sqrt"],
              'max_samples' : [.5,.9]}

scores = {'accuracy': 'accuracy' , 'precision':'precision', 'recall' : 'recall', 'f1':'f1'}
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
rf_val = RandomizedSearchCV(rf, rf_cv_params, cv=kfold, scoring = scores ,  refit='f1', n_jobs = -1, verbose = 1)

# Fit the model
rf_val.fit(X , y)
```

Best Hyperparameters

```python
#getting the best paramaters to be used with Random forest
rf_val.best_params_
```

#### Re-training the model with hyperparameters

```python
model = RandomForestClassifier(n_estimators=100,
                               min_samples_split=2,
                               min_samples_leaf=2,
                               max_samples=0.9,
                               max_features='sqrt',
                               max_depth=50)
```

```python
model.fit(X, y)
```

```python
import pickle
# Load the pickled model
with open('data/scaler_encoder_pipeline.pkl', 'rb') as file:
    pipeline = pickle.load(file)
```

## Evaluating on test dataset

```python
test_df = pd.read_csv("data/fraudTest.csv")
test_df.sample(2)
```

```python
columns_to_remove = ["Unnamed: 0" , "cc_num" , "first" , "last" , "street" , "lat" , "long", "zip" ,
                     "job" , "trans_num" , "unix_time" , "merch_lat" , "merch_long" ]
test_df.drop(columns=columns_to_remove, inplace=True)
```

```python
test_df['trans_date_trans_time'] = pd.to_datetime(test_df['trans_date_trans_time'])
test_df['dob'] = pd.to_datetime(test_df['dob'])
```

```python
test_df['trans_hour'] = test_df['trans_date_trans_time'].dt.hour
```

```python
def time_of_day(hour):
    if  0 <= hour <= 7:
        return "night"
    elif 8 <= hour <= 15:
        return "morning"
    else:
        return "midday"
        
# create the function
def age(age):
    if  2024 - age <= 29 :
        return "Young"
    elif 30 <= 2024 - age <=59 :
        return "Middle-Aged"
    else:
        return "Old"
```

```python
test_df['time_of_day'] = test_df['trans_hour'].apply(time_of_day)
```

```python
# extracting year from dob variable
test_df['year_birth']= test_df['dob'].dt.year
```

```python
test_df['age'] = test_df['year_birth'].apply(age)
```

```python
test_df.drop(columns=['trans_date_trans_time' , 'dob' , 'trans_hour' , 'year_birth'] ,inplace = True)
```

```python
X_test = test_df.drop(columns=['is_fraud'])
y_test = test_df['is_fraud']
```

```python
X_test = pd.DataFrame(pipeline.transform(X_test), columns=X_test.columns)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
y_pred = model.predict(X_test)
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Generate the classification report
report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report).transpose()
```

```python
# Plot the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
